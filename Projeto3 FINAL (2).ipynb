### Fatores socioeconômicos e demográficos podem influenciar a probabilidade de um jovem de 18 a 24 anos no Brasil estar na condição 'NemNem' (Nem estuda e Nem trabalha)?

# INTRODUÇÃO


No Brasil, uma parcela significativa de jovens entre 18 e 24 anos encontra-se na condição de não possuir uma posição no mercado de trabalho e no ambiente educacional, sendo comumente denominados "NemNem". Esses jovens frequentemente enfrentam vulnerabilidade social, sem acesso a oportunidades que poderiam ajudá-los a construir uma carreira ou melhorar seu nível de escolaridade. 

Esta situação é especialmente preocupante, pois a faixa etária de 18 a 24 anos é crucial para a transição para a vida adulta, onde os indivíduos começam a buscar independência financeira e a estruturar seu futuro profissional e educacional. Diversos fatores socioeconômicos e demográficos contribuem para essa condição, criando diferentes perfis de jovens que se encontram fora do mercado de trabalho e da educação. Entre os perfis mais comuns, destacam-se:

I. Jovens de baixa renda: Muitos jovens "NemNem" vêm de famílias de baixa renda, limitando seu acesso a recursos educacionais de qualidade e redes de apoio para conseguir emprego.

II. Desigualdade de gênero e raça: A condição "NemNem" afeta desproporcionalmente mulheres e jovens negros e pardos. Por exemplo, jovens mulheres frequentemente assumem responsabilidades domésticas ou cuidados familiares que as afastam do estudo e do trabalho.

III. Jovens em regiões com menos oportunidades: A condição "NemNem" é mais comum em regiões onde as oportunidades educacionais e de emprego são mais escassas, como no Norte e Nordeste do Brasil.

IV. Nível de escolaridade limitado: Jovens que interrompem os estudos precocemente, muitas vezes antes de concluírem o ensino médio, são mais suscetíveis a permanecerem fora do mercado de trabalho.

Com base nisso, este estudo sobre a condição dos jovens na categoria "NemNem" foi motivado pelo interesse em conhecer os fatores de influência deste grupo social.Os dados utilizados neste projeto foram extraídos da Pesquisa Nacional por Amostra de Domicílios (PNAD). Foram escolhidas características (**features**) julgadas necessárias pelo grupo estão sendo apresentadas na tabela abaixo, e a variável alvo (**target**) foi definida em uma coluna denominada "situação educacional-ocupacional". Esta coluna foi criada pelo grupo a partir das colunas de "força de trabalho potencial", "força de trabalho" e "pessoas que estão estudando no momento". Após a aplicação de alguns filtros, a coluna de situação educacional-ocupacional foi definida para o estudo.

# ANÁLISE EXPLORATÓRIA

Como a Target é uma variável **qualitativa**, optou-se por realizar uma análise exploratória detalhada dos dados utilizando ferramentas como: tabelas cruzadas, gráficos de barras, gráficos de área, gráficos de pizza e mapas de calor. Essas técnicas foram escolhidas devido às suas características específicas que permitem destacar diferentes aspectos das relações entre a Target e as Features escolhidas: Região, Sexo, Idade, Raça, Nível de Escolaridade e Renda Domiciliar Per Capita.

* **Gráficos de barras**: Foram utilizados para mostrar comparações diretas entre as categorias das Features, permitindo uma visualização clara das frequências relativas associadas a cada classe;

* **Gráficos de área**: Escolhidos para representar variações cumulativas ao longo de uma categoria contínua ou ordenada, como Idade, evidenciando tendências globais e destacando o impacto combinado de várias categorias;

* **Gráficos de pizza**: Aplicados para apresentar a distribuição percentual das categorias da Target, permitindo uma visualização imediata das proporções relativas dentro de uma variável qualitativa específica;

* **Mapas de Calor**: Utilizados para ilustrar as interações entre a Target e as Features em tabelas cruzadas, com intensidade de cores representando as frequências relativas. Isso permitiu identificar padrões e relações de forma intuitiva.

Para garantir comparabilidade e facilitar a interpretação dos resultados, utilizou-se a frequência relativa na normalização das análises, auxiliando na compreensão da influência de cada Feature na condição 'NemNem'."

## Análise Região

```python
# Tabela de contingência para 'Regiao' e 'SOE'
regiao_soe_ct = pd.crosstab(df.SOE,df.Regiao, normalize="columns")
regiao_soe_ct2 = pd.crosstab(df.Regiao,df.SOE, normalize="index")
regiao_soe_ct
```

```python
cores = ['#ffb3ba', "#ffdfba", "#baffc9", "#bae1ff"]

# Gráfico de Barras Empilhadas para 'Regiao' por 'SOE'
regiao_soe_ct2.plot(kind='bar', figsize=(8, 6), color=cores,stacked=True)
plt.title('Distribuição de Situação Ocupacional-Educacional (SOE) por Região')
plt.xlabel('Região')
plt.ylabel('Frequência relativa')
plt.legend(title="Região", bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.show()
```

```python
cores = ['#ffb3ba', "#ffdfba", "#ffffba", "#baffc9", "#bae1ff"]

# Configuração do gráfico
fig, axs = plt.subplots(2, 2, figsize=(18, 12)) 
# Achatar o array de eixos para iterar facilmente
axs = axs.flatten()

# Iterar sobre cada região e criar um gráfico de pizza
for ax, (soe, valores) in zip(axs[:len(regiao_soe_ct)], regiao_soe_ct.iterrows()):
    
# Criar o gráfico de pizza
    wedges, _ = ax.pie(valores, startangle=90, colors=cores, labels=None)
    
# Adicionar título
    ax.set_title(f'Situação ocupacional-educacional: {soe}')
    
# Calcular porcentagens
    porcentagens = [f'{valor / sum(valores) * 100:.1f}%' for valor in valores]
    
# Criar uma lista de legendas com categoria e porcentagem
    legend_labels = [f'{cat} - {pct}' for cat, pct in zip(valores.index, porcentagens)]
    
# Criar uma legenda ao lado de cada gráfico
    ax.legend(wedges, legend_labels, loc='center left', bbox_to_anchor=(1, 0.5), frameon=False)

# Remover os eixos extras que não estão sendo usados
for ax in axs[len(regiao_soe_ct):]:
    fig.delaxes(ax)

plt.tight_layout()
plt.show()
```

A análise dos dados revela uma relação significativa entre a condição de jovens que não estudam nem trabalham, os chamados "NemNem", e as diferentes regiões do Brasil. A maior concentração de indivíduos nessa situação encontra-se no Nordeste, seguido pela região Norte, enquanto as menores proporções estão no Sul e no Sudeste. Essa diferença sugere que fatores regionais, como as condições econômicas, oportunidades educacionais e de emprego, desempenham um papel determinante no comportamento desse grupo.

O Nordeste apresenta a maior proporção de NemNem em comparação às outras regiões. Isso reflete os desafios históricos e estruturais da região, como altas taxas de desemprego, menor acesso à educação de qualidade e desigualdade socioeconômica persistente. O Norte, por sua vez, também apresenta proporções elevadas de NemNem, reforçando o impacto das desigualdades regionais.

Em contrapartida, as regiões Sul e Sudeste apresentam as menores proporções de jovens NemNem, o que indica que viver nessas áreas reduz a chance de pertencer a esse grupo. Isso pode ser atribuído ao maior dinamismo econômico, com mais oportunidades de trabalho formal, além de uma infraestrutura educacional mais consolidada nessas regiões. Já o Centro-Oeste apresenta uma posição intermediária, com proporções moderadas de NemNem, o que sugere uma situação menos crítica, mas ainda com desafios pontuais.

A análise evidencia a influência direta das desigualdades regionais sobre a exclusão educacional e ocupacional. A situação dos NemNem é mais prevalente onde há menor oferta de oportunidades de trabalho e educação, enquanto regiões mais desenvolvidas economicamente conseguem minimizar essa exclusão.

### Criando DataFrame rebalanceado

```python
# Filtrando e equilibrando as classes
df_class_0 = df[df['y'] == 0]
df_class_1 = df[df['y'] == 1]

min_class_count = min(len(df_class_0), len(df_class_1))
df_class_0_balanced = df_class_0.sample(n=min_class_count, random_state=42)
df_class_1_balanced = df_class_1.sample(n=min_class_count, random_state=42)

df_balanced = pd.concat([df_class_0_balanced, df_class_1_balanced])

Xb = df_balanced[['Regiao', 'Sexo', 'Idade', 'Raca', 'NE', 'RFCP']]
yb = df_balanced['y']

# Convertendo variáveis categóricas para dummies e garantindo o tipo float
Xb = pd.get_dummies(Xb, columns=['Regiao', 'Sexo', 'Raca', 'NE', 'RFCP'], drop_first=True).astype(float)

# Divisão entre treino e teste
Xb_train, Xb_test, yb_train, yb_test = train_test_split(Xb, yb, test_size=0.33, random_state=42)
```

# MODELO 1: Regressão logística

A regressão logística é uma técnica estatística amplamente utilizada para prever a probabilidade de ocorrência de um evento binário com base em variáveis independentes. No contexto do projeto, ela pode ser aplicada para prever a condição "NemNem" (não estuda, nem trabalha) entre jovens de 18 a 24 anos.

## Como Funciona

A regressão logística transforma as probabilidades em uma escala logarítmica (logit) e estima os coeficientes das variáveis independentes (\( \beta \)) usando o método de máxima verossimilhança (MLE). A fórmula básica do modelo é:

$$
\text{logit}(\pi) = \ln\left(\frac{\pi}{1-\pi}\right) = \beta_0 + \beta_1X_1 + \dots + \beta_kX_k
$$

### Definições

$\text( \pi )$: Probabilidade prevista de o jovem estar na condição "NemNem".


$\text( x )$: Variáveis independentes, como Região, Sexo, Idade, entre outras.


$\text( \beta )$: Coeficientes que medem o impacto de cada variável independente na chance de o jovem ser "NemNem".


## Interpretação no Contexto do Projeto

Os coeficientes $( \beta )$ podem ser interpretados como a influência de cada variável independente. Para facilitar a compreensão, eles são frequentemente exponenciados, convertendo-os em *razões de chance (OR)*:

- $( \text{OR} > 1 )$: O aumento da variável está associado a uma maior chance de o jovem ser "NemNem".
- $( \text{OR} < 1 )$: O aumento da variável está associado a uma menor chance de o jovem ser "NemNem".

Por exemplo, se a razão de chance para mulheres for 1,5 em relação aos homens, isso indica que as mulheres têm 1,5 vezes mais chance de estarem na condição "NemNem", mantendo todas as outras variáveis constantes.

_link: https://www.ibm.com/br-pt/topics/logistic-regression_

# Modelo 2: Random Forest Classifier

O *Random Forest* é um algoritmo de aprendizado supervisionado que combina múltiplas árvores de decisão para realizar tarefas de classificação ou regressão. Ele é amplamente utilizado devido à sua capacidade de lidar com conjuntos de dados complexos, balancear vieses e variâncias, e evitar o overfitting.

## Como Funciona

1. *Construção de Árvores*:  
   O algoritmo cria diversas árvores de decisão a partir de diferentes amostras aleatórias do conjunto de dados (bootstrap).

2. *Divisão Aleatória de Features*:  
   Em cada nó da árvore, um subconjunto aleatório de features é considerado para determinar a melhor divisão, promovendo diversidade entre as árvores.

3. *Combinação de Resultados*:  
   - Para classificação: O algoritmo utiliza o voto majoritário das árvores para determinar a classe final.
   - Para regressão: A média das previsões de todas as árvores é usada como resultado final.

## Vantagens
- Lida bem com dados com muitas variáveis e correlações.
- Reduz o risco de overfitting ao combinar múltiplas árvores.
- É robusto contra ruído nos dados e outliers.

_link: https://didatica.tech/o-que-e-e-como-funciona-o-algoritmo-randomforest/_

# Modelo 3: Gradient Boosting Classifier

O Gradient Boosting Classifier é um modelo de aprendizado supervisionado amplamente utilizado para tarefas de classificação. Ele constrói um conjunto de classificadores fracos (como árvores de decisão) de maneira sequencial, com cada modelo corrigindo os erros do anterior, a fim de minimizar uma função de perda. Este método é conhecido por sua alta precisão e flexibilidade em modelar dados complexos.

## Como Funciona

- Inicialização com um Modelo Simples:
O algoritmo começa com um modelo inicial, geralmente uma previsão constante baseada na média ou moda dos rótulos.

- Construção Sequencial:
Cada árvore de decisão subsequente é treinada para corrigir os erros residuais do modelo anterior.

- Ajuste por Gradiente:
O modelo minimiza iterativamente uma função de perda, como a log-loss para classificação, ajustando os pesos dos dados e otimizando os novos modelos.

- Combinação Final:
As previsões de todas as árvores são somadas para obter a classificação final.

## Vantagens

Lida bem com dados desbalanceados.
Pode ser ajustado para minimizar diferentes funções de perda, o que o torna flexível para diversos problemas.
Tem bom desempenho em conjuntos de dados de tamanhos pequenos a médios.

```python
# Criando o modelo Gradient Boosting
modelo_gbm = GradientBoostingClassifier(n_estimators=100, random_state=42)

# Treinando o modelo
modelo_gbm.fit(X_train, y_train)

# Fazendo previsões no conjunto de teste
y_pred = modelo_gbm.predict(X_test)

# Avaliando o desempenho do modelo
acuracia = accuracy_score(y_test, y_pred)
print(f"Acurácia: {acuracia * 100:.2f}%")

# Matriz de Confusão
matriz_confusao = confusion_matrix(y_test, y_pred)
print("Matriz de Confusão:")
print(matriz_confusao)

print("Relatório de Classificação:")
relatorio = classification_report(y_test, y_pred)
print(relatorio)
```

Comparativamente, o *Random Forest Classifier* exibiu robustez e versatilidade, com boa precisão para a classe 0 e um desempenho razoável para a classe 1 (Ftdes, Ftoc, FTP), oferecendo ainda uma vantagem adicional na avaliação da importância das variáveis. Já a *Regressão Logística*, embora mais simples e intuitiva, apresentou um desempenho ligeiramente inferior, com acurácia de 74% e menor recall, mas compensou ao oferecer estabilidade e transparência nos coeficientes explicativos. Assim, o Gradient Boosting emergiu como a melhor escolha para este problema, especialmente quando a priorização da classe "NemNem" é essencial.
